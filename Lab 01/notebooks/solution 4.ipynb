## Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt


def histogram(df: str, name: str) -> None:
    """
    Create and display a histogram with KDE (Kernel Density Estimation) line for a given DataFrame column.

    Parameters:
        df (str): The DataFrame containing the data.
        name (str): The name of the column to create the histogram for.

    Returns:
        None

    Example:
        histogram(my_dataframe, 'column_name')
    """
    plt.figure(figsize=(10, 10))
    sns.histplot(data=df, x=name, kde=True)
    plt.title(f"Histogram of {name}")
    plt.show()


def pieplot(df: str, name: str) -> None:
    """
    Create and display a pie chart for the distribution of unique values in a given DataFrame column.

    Parameters:
        df (str): The DataFrame containing the data.
        name (str): The name of the column to create the pie chart for.

    Returns:
        None

    Example:
        pieplot(my_dataframe, 'column_name')
    """
    plt.figure(figsize=(10, 10))
    plt.pie(
        df[name].value_counts(),
        labels=df[name].value_counts().index,
        autopct="%1.1f%%",
        startangle=140,
    )
    plt.title(f"Pie Plot of {name}")
    plt.show()


# Identify numerical columns in the DataFrame 'wine_df'
wine_numberial_columns = wine_df.columns[wine_df.dtypes != "object"]

# Identify numerical columns in the DataFrame 'housedf'
house_numberical_columns = housedf.columns[housedf.dtypes != "object"]

# Identify numerical columns in the DataFrame 'wine_df'
wine_categorical_columns = wine_df.columns[wine_df.dtypes == "object"]

# Identify numerical columns in the DataFrame 'housedf'
house_categorical_columns = housedf.columns[housedf.dtypes == "object"]
```

### Line Plot

```python
plt.figure(figsize=(10, 10))
sns.lineplot(wine_df, x="fixed acidity", y="quality", hue="type")
plt.title("Acidity VS Quality")
plt.show()
```

### Heatmap

```python
plt.figure(figsize=(20, 10))
sns.heatmap(
    wine_df[wine_numberial_columns].corr(),
    annot=True,
    fmt=".2f",
    square=True,
    cmap="Blues_r",
)
plt.title("Wine Dataset Correlation", size=15)
plt.show()
```
