### Import required packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

### K Means Clustering

```python
# given sample data
X = np.array([(1, 4), (1, 3), (0, 4), (5, 1), (6, 2), (4, 0)])

# initialize centroids
centroid1, centroid2 = X[0], X[1]

# create an array to store cluster assignments
clusters = np.zeros(len(X))

# perform k-means clustering for 5 iterations
for _ in range(5):
    for i in range(len(X)):
        distance_to_centroid1 = np.linalg.norm(X[i] - centroid1)
        distance_to_centroid2 = np.linalg.norm(X[i] - centroid2)
        clusters[i] = (
            0 if distance_to_centroid1 < distance_to_centroid2 else 1
        )  # Assign to cluster with the closest centroid
    # Update centroids based on the mean of the points in each cluster
    centroid1 = np.mean(X[clusters == 0], axis=0)
    centroid2 = np.mean(X[clusters == 1], axis=0)

print("Labels for each point after k-means clustering: ")
for i, label in enumerate(clusters):
    print(f"Point {X[i]} is assigned to Cluster {int(label)}")
```

### 3D ScatterPlot

```python
# 3D plot
fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(111, projection="3d")

# Scatter plot for Cluster 1
ax.scatter(
    X[clusters == 0][:, 0],
    X[clusters == 0][:, 1],
    np.zeros_like(X[clusters == 0][:, 0]),
    c="red",
    label="Cluster 1",
    s=100,
)

# Scatter plot for Cluster 2
ax.scatter(
    X[clusters == 1][:, 0],
    X[clusters == 1][:, 1],
    np.zeros_like(X[clusters == 1][:, 0]),
    c="blue",
    label="Cluster 2",
    s=100,
)

# Scatter plot for centroids
ax.scatter(
    centroid1[0], centroid1[1], 0, marker="d", s=150, c="green", label="Centroid 1"
)
ax.scatter(
    centroid2[0], centroid2[1], 0, marker="d", s=150, c="green", label="Centroid 2"
)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.title("K-Means Clustering (3D)")
plt.legend()
plt.show()
```
